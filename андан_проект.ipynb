## 1ый датасет

[Данный датасет](https://www.kaggle.com/datasets/thomasnibb/amsterdam-house-price-prediction) - информация о домах на продаже в Амстердаме. Рассмотрим столбцы и преобразуем так, как нам надо.

**Address**: полный адрес. Эта информация полезная, но мы будем разбивать данный столбец на два: location (регион) и country (страна). 

**Zip**: почтовый индекс. Стоит удалить, так как эта информация не будет являться важным признаком, и не во всех других датасетах есть такая информация, а заполнить пропуски не получится. 

**Price**: цена(в евро). Это наш таргет. Переведем в доллары.

**Area**: площадь(в метрах).

**Room**: количество комнат.

**Lon**: долгота. Пока что оставим, может пригодиться.

**Lat**: широта. Аналогично.

```python
df_amsterdam = pd.read_csv('amsterdam.csv')

df_amsterdam[['location', 'country']] = df_amsterdam['Address'].str.split(',', expand=True)

df_amsterdam.drop(columns=['Unnamed: 0', 'Address', 'Zip'], inplace=True)

df_amsterdam['Price'] = df_amsterdam['Price'] * 1.1

df_amsterdam['Area'] = df_amsterdam['Area'].astype(float)

df_amsterdam
```

## 2ой датасет

[Данные датасеты](https://www.kaggle.com/datasets/saipavansaketh/pune-house-data?select=Bangalore++house+data.csv) из соревнования Kaggle с информацией о домах на продаже в Индии. Рассмотрим столбцы и преобразуем так, как нам надо.

**Area**: площадь(в футах). Переведем в метры.

**BHK**: Количество спален, а также 1 холл и 1 кухня.

**Bathroom**: Количество ванных комнат.

**Locality**: Населенный пункт, в котором находится недвижимость.

**Parking**: Количество мест для парковки.

**Price**: цена(в индийских рупиях). таргет.

```python
df_delhi = pd.read_csv('delhi.csv')

df_delhi.drop(columns=['Furnishing', 'Status', 'Transaction', 'Type', 'Per_Sqft'], inplace=True)
df_delhi['bedrooms'] = 0
df_delhi['kitchen'] = 0
df_delhi['rooms'] = 0
df_delhi.loc[pd.to_numeric(df_delhi['BHK']) > 0, 'bedrooms'] = pd.to_numeric(df_delhi['BHK'])
df_delhi.loc[pd.to_numeric(df_delhi['BHK']) > 0, 'kitchen'] = 1
df_delhi.loc[pd.to_numeric(df_delhi['BHK']) > 0, 'rooms'] = pd.to_numeric(df_delhi['BHK']) + pd.to_numeric(df_delhi['Bathroom']) + 2

df_delhi.drop(columns=['BHK'], inplace=True)

df_delhi['country'] = 'India'

df_delhi.rename(columns={'rooms' : 'Room', 'Bathroom' : 'Bathrooms', 'bedrooms' : 'Bedrooms', 'kitchen' : 'Kitchens', 'Locality' : 'location'}, inplace=True)

df_delhi['Area'] = pd.to_numeric(df_delhi['Area']) * 0.0929
df_delhi['Price'] = df_delhi['Price'] * 0.012177

df_delhi['Area'] = df_delhi['Area'].astype(float)

df_delhi
```

# Предварительная обработка 

Посмотрим на нашу табличку и на количество пропущенных значений в столбцах.

```python
df = df.reset_index()
df.drop(columns=['index'],inplace=True)
df
```

```python
df.info()
```

Посмотрим на количество пропусков в данных и подумаем, как и заполнить.

```python
df.isna().sum()
```

Теперь глянем на распределение количества пропусков по столбцам.

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])

percent.plot(kind="bar", figsize = (8,6), fontsize = 10)
plt.xlabel("Column", fontsize = 20)
plt.ylabel("Percent", fontsize = 20)
plt.title("Пропуски", fontsize = 20)
```

В Parking и Kitchens очень много пропусков. Просто удалим эти столбцы.

```python
df.drop(columns=['Parking'],inplace=True)
df.drop(columns=['Kitchens'],inplace=True)
```

Строки в Price с пропусками сразу удаляем, так как это таргет, и если они попадут в тренировочную выборку, то мы получим ошибку.

```python
df.drop(index=df[df['Price'].isna() == True].index, inplace=True)
```

Со строками с пропусками Area_lot (площадь земельного участка) будем работать следующим образом: заполним их значением Area (площадь жилого помещения).

```python
df['Area'] = df['Area'].astype(float)
```

```python
df['Area_lot'] = df['Area_lot'].fillna(df['Area'])
```

Пропуски в столбце Floors заполним 1, так как при большем количестве этажей это было бы указано, или же мы такие ситуации обработали в предыдущем пункте. Аналогично поступаем с пропусками в Bedrooms и Bathrooms.


```python
df['Floors'] = df['Floors'].fillna(1)
df['Bedrooms'] = df['Bedrooms'].fillna(1)
df['Bathrooms'] = df['Bathrooms'].fillna(1)
```

Room с пропусками заполним суммой комнат (ванны, спальни).

```python
df['Room'] = df['Room'].fillna(df['Bathrooms'].astype(float) + df['Bedrooms'].astype(float))
```

Пропуски в сountry и location заполним константной строкой (можно пустной, но мы используем 'another' (это не сыграет роли, так как будет использоваться, как категориальный признак)), так как узнать эти данные не получится.

```python
df['country'] = df['country'].fillna('another')
df['location'] = df['location'].fillna('another')
```

Age заполним средним столбца, чтобы данные пропуски не сильно влияли на результат, но при этом учитывались, а Age_new заполним 0, так как это значит, что дом не был на ремонте, как во многих данных.

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age_new'] = df['Age_new'].fillna(0)
```

Для получения координат будем использовать данные location и country, которые у нас есть и функцию geocode из модуля geopandas.

```python
df['full_address'] = df.location + ',' + df.country
df.loc[(df['location'] == 'another'), 'full_address'] = df['country']
df.loc[(df['country'] == 'another'), 'full_address'] = df['location']
df.loc[((df['location'] == 'another') & (df['country'] == 'another')), 'full_address'] = ''
```

```python
for place in df[df['Lon'].isna() == True]['full_address'].unique():
    place_short = place
    if len(place) > 70:
        place_short = place.split(',')[-1]
    coord = geocode(place_short, timeout=10)
    if not coord.address.isna()[0]:
        df.loc[((df['Lat'].isna() == True) & (df['full_address'] == place) & (df['Lat'].isna() == True)), 'Lat'] = coord.geometry.iloc[0].x
        df.loc[((df['Lon'].isna() == True) & (df['full_address'] == place) & (df['Lon'].isna() == True)), 'Lon'] = coord.geometry.iloc[0].y
    else:
        df.loc[((df['Lat'].isna() == True) & (df['full_address'] == place) & (df['Lat'].isna() == True)), 'Lat'] = 0
        df.loc[((df['Lon'].isna() == True) & (df['full_address'] == place) & (df['Lon'].isna() == True)), 'Lon'] = 0
```

# Создание новых признаков

Создадим признак, который будет отвечать за то, сколько процентнов относительно всей площади жилого помещения составляют спальни и ванные. 

Также создадим признаки, которые отвечают соответственно за суммарную площадь отведенную под спальни и под ванные.

```python
df['PercentBeds'] = 0
df.loc[df['Room'] > 0, 'PercentBeds'] = (df['Bedrooms'] * 100 / df['Room'])
df['PercentBaths'] = 0
df.loc[df['Room'] > 0, 'PercentBaths'] = (df['Bathrooms'] * 100 / df['Room'])

df['AreaBeds'] = 0
df.loc[df['Room'] > 0, 'AreaBeds'] = (df['Area']  * df['Bedrooms'] / df['Room'])
df['AreaBaths'] = 0
df.loc[df['Room'] > 0, 'AreaBaths'] = (df['Area'] * df['Bathrooms'] / df['Room'])
```

И создадим более сложные признаки для улучшения работы, например, линейной регрессии, признаки, которые будут равны квадрату, кубу и корню признаком количества спален, ванн, комнат и площадей.

```python
df["Bedrooms2"] = df["Bedrooms"] ** 2
df["Bedrooms3"] = df["Bedrooms"] ** 3
df["BedroomsSq"] = np.sqrt(df["Bedrooms"])  

df["Bathrooms2"] = df["Bathrooms"] ** 2
df["Bathrooms3"] = df["Bathrooms"] ** 3
df["BathroomsSq"] = np.sqrt(df["Bathrooms"])

df["Room2"] = df["Room"] ** 2
df["Room3"] = df["Room"] ** 3
df["RoomSq"] = np.sqrt(df["Room"])

df["Area2"] = df["Area"] ** 2
df["Area3"] = df["Area"] ** 3
df["AreaSq"] = np.sqrt(df["Area"])  

df["Area_lot2"] = df["Area_lot"] ** 2
df["Area_lot3"] = df["Area_lot"] ** 3
df["Area_lotSq"] = np.sqrt(df["Area_lot"])  
```

# Гипотезы

## Первая гипотеза

Из графика распределения средних цен по странам мы могли заметить, что средняя цена на дом в США находится примерно на середине графика. То есть мы можем выдвинуть следующую нулевую гипотезу:

средняя цена на дом в США равна усредненной цене на дома по остальным странам (по оставшейся части датасета).

Таким образом мы выдвигаем гипотезы:

$H_0: mean(Price_{USA}) = mean(Price_{another-country})$

$H_1: mean(Price_{USA}) != mean(Price_{another-country})$

Данные независимы, и мы можем использовать t-тест или же z-тест со значением value=0.

```python
price_usa = df[df['country'] == 'USA']['Price']
price_another = df[df['country'] != 'USA']['Price']
```

Также для проведения теста надо посмотреть на соотношение дисперсий (должно быть меньше, чем 4:1).

```python
print(np.var(price_usa) / np.var(price_another))
```

```python
print(np.var(price_another) / np.var(price_usa))
```

Все подходит. Можем проверять.

```python
sts.ttest_ind(price_usa, price_another)
```

```python
wsts.ztest(price_usa, price_another)
```

Поскольку мы не меняли параметр equal_var при проведении t-теста, он по умолчанию равен True, что говорит о том, что проверялась нулевая гипотеза.

Также при проведении z-теста мы оставили значение 0 (по умолчанию), из-за чего проверялась нулевая гипотеза.

Итог: при проведении двух тестов мы получили pvalue = 0.03 (по умолчанию берем уровень значимости $\alpha$ = 0.05), что меньше $\alpha$, а значит мы отвергаем нулевую гипотезу $\Rightarrow$ $mean(Price_{USA}) != mean(Price_{another})$.

## Вторая гипотеза

Также из того же графика можно заметить, что средняя цена на дома в Иране и в датасете без указания страны были примерно равны.

Проверим тогда такие гипотезы:

$H_0: mean(Price_{Iran}) = mean(Price_{another})$

## Третья гипотеза

Можно попробовать выполнить анализ Уэлча, чтобы посмотреть на равенство по двум столбцам. 

Например, выдвинем следующие гипотезы:

$H_0:$ для каждого значения $Floors$ равны $Price$

$H_1:$ для каждого значения $Floors$ не равны $Price$

Для начала надо провести тест Бартлетта для проверки равенства дисперсий между группами.

```python
all = [np.array(df[df['Floors'] == count_floors]['Price']) for count_floors in df['Floors'].unique()]
```

```python
sts.bartlett(all[0], all[1], all[2], all[3], all[4], all[5])
```

Соответственно получаем pvalue близкое к 0, что отвергает гипотезу о равенстве дисперсий, и мы можем переходить к анализу Уэлча.

```python
pg.welch_anova(dv='Price', between='Floors', data=df)
```

Получаем pvalue (почти 0) меньшее $\alpha$ $\Rightarrow$ гипотеза о равенстве цен отклоняется.

Также можно посмотреть на равенство между каждой группой:

```python
z = pg.pairwise_gameshowell(dv='Price', between='Floors', data=df)
z['pval'] = z['pval'].round(10)
z
```

Можно заметить, что между некоторыми группами все же нельзя отклонить нулевую гипотезу о равенстве, однако в сумме они являются меньшинством.

## Подготовка

## Ridge-регрессия

Обучим обычную Ridge-регрессию на признаках, которые были изначально (без искусственно созданных).

Будем использовать One-Hot Encoding для категориальных признаков.

```python
notnum_feats = ['country', 'location', 'full_address']
num_feats = ['Bedrooms', 'Bathrooms', 'Area', 'Area_lot', 'Floors', 'Age', 
                'Age_new', 'Lat', 'Lon', 'Room']
```

```python
X_train_now = X_train[notnum_feats + num_feats]
X_test_now = X_test[notnum_feats + num_feats]
```

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), notnum_feats),
])

pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', Ridge())
])

model = pipeline.fit(X_train_now, y_train)
y_pred = model.predict(X_test_now)
y_train_pred = model.predict(X_train_now)

print("Test RMSE = %.4f" % mean_squared_error(y_test, y_pred, squared=False))
print("Train RMSE = %.4f" % mean_squared_error(y_train, y_train_pred, squared=False))

print("Test R^2 = %.4f" % r2_score(y_test, y_pred))
print("Train R^2 = %.4f" % r2_score(y_train, y_train_pred))
```

Теперь попробуем отнормировать численные признаки, чтобы качество модели улучшилось.

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), notnum_feats),
    ('scaling', StandardScaler(), num_feats)
])

pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', Ridge())
])

model = pipeline.fit(X_train_now, y_train)
y_pred = model.predict(X_test_now)
y_train_pred = model.predict(X_train_now)

print("Test RMSE = %.4f" % mean_squared_error(y_test, y_pred, squared=False))
print("Train RMSE = %.4f" % mean_squared_error(y_train, y_train_pred, squared=False))

print("Test R^2 = %.4f" % r2_score(y_test, y_pred))
print("Train R^2 = %.4f" % r2_score(y_train, y_train_pred))
```

На тестовой выборке получили меньшую ошибку RMSE, что означает следущее: модель предсказывает цену дома с ошибкой в то количество едениц, чему равна метрика. 

Метрика R^2 дала значение чуть больше половины, но все равно мало. Модель считается идеальной по этой метрике, когда значение стремится к 1. 

Теперь обучим такую же регрессию, но с новыми признаками.

```python
notnum_feats = ['country', 'location', 'full_address']
num_feats = ['Bedrooms', 'Bathrooms', 'Area', 'Area_lot', 'Floors', 'Age', 
                'Age_new', 'Lat', 'Lon', 'Room',
                'PercentBeds', 'PercentBaths', 'AreaBeds', 'AreaBaths', 'Bedrooms2',
                'Bedrooms3', 'BedroomsSq', 'Bathrooms2', 'Bathrooms3', 'BathroomsSq',
                'Room2', 'Room3', 'RoomSq', 'Area2', 'Area3', 'AreaSq', 'Area_lot2',
                'Area_lot3', 'Area_lotSq']
```

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), notnum_feats),
    ('scaling', StandardScaler(), num_feats)
])

pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', Ridge())
])

model = pipeline.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)

print("Test RMSE = %.4f" % mean_squared_error(y_test, y_pred, squared=False))
print("Train RMSE = %.4f" % mean_squared_error(y_train, y_train_pred, squared=False))

print("Test R^2 = %.4f" % r2_score(y_test, y_pred))
print("Train R^2 = %.4f" % r2_score(y_train, y_train_pred))
```

Качество с добавленными признаками улучшилось.

Теперь RMSE даёт ошибку еще меньше. Уже получше. Значит созданные признаки улучшают модель.

R^2 также улучшилось.

## Lasso-регрессия

Теперь попробуем обучить Lasso-регрессию на всех признаках.

## Логистическая регрессия

Попробуем теперь ввести классификацию дороговизны дома таким образом, что он будет равен 1 в тех случаях, когда цена дома больше среднего значения таргета, и равен 0 - меньше.  

Оценивать будем с помощью PR-AUC и AUC-ROC, чтобы избавиться от проблемы возможной неравности классов. 

```python
mean_value = y_train.mean()
```

```python
y_train_log = pd.DataFrame(y_train > mean_value)
y_test_log = pd.DataFrame(y_test > mean_value)

y_train_log.loc[y_train_log['Price'] == True, 'Price'] = 1
y_train_log.loc[y_train_log['Price'] == False, 'Price'] = 0
y_train_log = y_train_log['Price']
y_train_log = y_train_log.astype('int')

y_test_log.loc[y_test_log['Price'] == True, 'Price'] = 1
y_test_log.loc[y_test_log['Price'] == False, 'Price'] = 0
y_test_log = y_test_log['Price']
y_test_log = y_test_log.astype('int')
```

```python
column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), notnum_feats),
    ('scaling', StandardScaler(), num_feats)
])

pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', LogisticRegression(max_iter=100000))
])

model = pipeline.fit(X_train, y_train_log)

precision, recall, _ = precision_recall_curve(y_test_log, model.decision_function(X_test))
print('Test PR-AUC: %.5f' %auc(recall, precision))
print('Test AUC-ROC: %.5f' % roc_auc_score(y_test_log, model.decision_function(X_test)))

precision, recall, _ = precision_recall_curve(y_train_log, model.decision_function(X_train))
print('Train PR-AUC: %.5f' %auc(recall, precision))
print('Train AUC-ROC: %.5f' % roc_auc_score(y_train_log, model.decision_function(X_train)))
```

Тут мы получили очень неплохие значения метрик.

AUC-ROC получился около 0.93, что довольно близко к 1. Это хорошо для данной метрики.

Значение PR-AUC мы получили близко 0.9, что также означает хорошее предсказание.
